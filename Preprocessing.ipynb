```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from scipy.stats import skew, kurtosis

# Load the dataset
data = pd.read_csv('UCI.csv')

# Drop non-numeric columns (if 'Activity' is the only non-numeric one)
X = data.drop('Activity', axis=1)
y = data['Activity']

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Feature Engineering
def generate_features(df):
    feature_df = pd.DataFrame()

    # Mean, Std, Min, Max
    feature_df['mean'] = df.mean(axis=1)
    feature_df['std'] = df.std(axis=1)
    feature_df['min'] = df.min(axis=1)
    feature_df['max'] = df.max(axis=1)

    # Skewness and Kurtosis
    feature_df['skewness'] = df.apply(skew, axis=1)
    feature_df['kurtosis'] = df.apply(kurtosis, axis=1)

    # Signal Magnitude Area (SMA) - useful for accelerometer/gyroscope data
    feature_df['sma'] = df.abs().sum(axis=1)

    # Root Mean Square (RMS)
    feature_df['rms'] = np.sqrt(np.mean(df**2, axis=1))

    # Energy (sum of squares)
    feature_df['energy'] = np.sum(df**2, axis=1)

    return feature_df

# Generate new features
feature_engineered_data = generate_features(X)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(feature_engineered_data)

# Convert back to DataFrame for clarity
X_final = pd.DataFrame(X_scaled, columns=feature_engineered_data.columns)

# Display the new feature set
X_final.head()
```

```python
import numpy as np

# FFT Feature Engineering
def generate_fft_features(df):
    fft_features = pd.DataFrame()

    # Compute FFT for each row
    fft_values = np.fft.fft(df, axis=1)
    fft_magnitude = np.abs(fft_values)

    # FFT-Based Features
    fft_features['fft_mean'] = np.mean(fft_magnitude, axis=1)
    fft_features['fft_std'] = np.std(fft_magnitude, axis=1)
    fft_features['fft_max'] = np.max(fft_magnitude, axis=1)
    fft_features['fft_min'] = np.min(fft_magnitude, axis=1)

    # Dominant Frequency (peak frequency)
    fft_features['dominant_freq'] = np.argmax(fft_magnitude, axis=1)

    return fft_features

```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from scipy import stats

# Load the dataset
df = pd.read_csv("UCI.csv")

# Step 1: Handle Missing Values
# Separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = SimpleImputer(strategy='mean').fit_transform(df[numeric_cols])

categorical_cols = df.select_dtypes(exclude=[np.number]).columns
df[categorical_cols] = SimpleImputer(strategy='most_frequent').fit_transform(df[categorical_cols])

# Step 2: Encode Categorical Variables
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Step 3: Remove Outliers using Z-score
z_scores = np.abs(stats.zscore(df[numeric_cols]))
df = df[(z_scores < 3).all(axis=1)]

# Step 4: Scale Features
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Step 5: Remove Duplicates
df.drop_duplicates(inplace=True)

print("Preprocessing complete. Dataset shape:", df.shape)

```

```python
# Save the preprocessed dataset
df.to_csv("UCI_preprocessed.csv", index=False)
print("Preprocessed dataset saved as 'UCI_preprocessed.csv'")

```
